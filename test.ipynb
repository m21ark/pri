```python
import re 

def convert_to_years(value):
    # Regular expressions to match different formats
    months_pattern = r'(\d+)\s*months?'
    weeks_pattern = r'(\d+)\s*weeks?'
    days_pattern = r'(\d+)\s*days?'
    years_pattern = r'(\d+)\s*years?'

    # Check for specific formats and convert to years
    if isinstance(value, str):
        if re.match(months_pattern, value):
            match = re.match(months_pattern, value)
            months = int(match.group(1))
            return months / 12.0
        elif re.match(weeks_pattern, value):
            match = re.match(weeks_pattern, value)
            weeks = int(match.group(1))
            return weeks / 52.0
        elif re.match(days_pattern, value):
            match = re.match(days_pattern, value)
            days = int(match.group(1))
            return days / 365.0
        elif re.match(years_pattern, value):
            match = re.match(years_pattern, value)
            years = int(match.group(1))
            return years
        else:
            return np.nan
    else:
        return np.nan
        

    
```

```python
converted_values = [convert_to_years(value) for value in list_age_weaning]

# Print the converted values
for i, value in enumerate(converted_values):
    print(f"{list_age_weaning[i]:<50} => {value:.2f} years")
```
